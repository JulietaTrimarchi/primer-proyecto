```python
# Importo librería a utilizar

import pandas as pd
```

```python
# Abro Dataset Movies

dataset_movies = pd.read_csv("C:/Users/Juli/Desktop/Data Science/Proyecto individual I Machine Learning/Datasets originales/movies_dataset.csv",
                                delimiter=',',
                                quotechar='"',
                                escapechar='\\',
                                low_memory = False)
```

```python
#Informacion del Dataset

dataset_movies.info()
```

```python
dataset_movies.head()
```

```python
# Convierto los datos de tipo objeto a flotante de la columna budget

dataset_movies['budget'] = pd.to_numeric(dataset_movies['budget'], errors='coerce')
dataset_movies["budget"]
```

```python
# Transformo nulos en "0" en columnas budget y revenue
dataset_movies[['budget', 'revenue']] = dataset_movies[['budget', 'revenue']].fillna(0)
```

```python
# Verifico paso anterior
print(dataset_movies[['budget', 'revenue']].isnull().sum())
```

```python
# Convierto a formato datetime la columna release_date

dataset_movies['release_date'] = pd.to_datetime(dataset_movies['release_date'], errors='coerce', format='%Y-%m-%d')
dataset_movies["release_date"].info()
```

```python
# Transformo las fechas a formato AAAA-mm-dd
dataset_movies['release_date'] = dataset_movies['release_date'].dt.strftime('%Y-%m-%d')
```

```python
# Elimino los valores nulos
dataset_movies = dataset_movies.dropna(subset=['release_date'])
```

```python
# Verifico
print(dataset_movies['release_date'].isnull().sum())
```

```python
# Creo columna que contiene el año de estreno de cada película
dataset_movies['release_year'] = pd.to_datetime(dataset_movies['release_date']).dt.year
```

```python
# Importo numpy

import numpy as np
```

```python
# Creo la columna return con el cociente entre revenue y budget

dataset_movies['return'] = np.where(dataset_movies['budget'] == 0, 0, dataset_movies['revenue'] / dataset_movies['budget'])
```

```python
dataset_movies[["revenue","budget","return"]].head()
```

```python
# Elimino columnas que no usaré
dataset_movies = dataset_movies.drop(['video','imdb_id','adult','original_title','poster_path','homepage'], axis=1)
```

```python
# Chequeo cantidad de columnas

num_columns = dataset_movies.shape[1]
print(num_columns)
```

```python
# Importo libreria que me será util 

import calendar
```

```python
# Agrego una columna con los meses de estreno de cada película

dataset_movies['release_month'] = pd.to_datetime(dataset_movies['release_date']).dt.month
dataset_movies
```

```python
# Creo un diccioario para los nombres de los meses en español

meses_español = {   1: 'enero',
                    2: 'febrero',
                    3: 'marzo',
                    4: 'abril',
                    5: 'mayo',
                    6: 'junio',
                    7: 'julio',
                    8: 'agosto',
                    9: 'septiembre',
                    10: 'octubre',
                    11: 'noviembre',
                    12: 'diciembre'
}
```

```python
dataset_movies['name_month'] = dataset_movies['release_month'].map(meses_español)
dataset_movies
```

```python
print(dataset_movies['release_year'].min())
print(dataset_movies['release_year'].max())
```

```python
# Creo una copia del dataframe

data_mes = dataset_movies.copy()
```

```python
# Voy a realizar un filtro sobre el mismo para obtener las columnas titulo y mes de estreno.

data_mes = data_mes[['title', 'name_month']]
data_mes
```

```python
# Voy a crear un Dataframe que contenga el mes y el total de películas estrenadas

# Conteo de películas por mes
conteo_peliculas = data_mes.groupby('name_month').size().reset_index(name='total_movies')


# Lista de meses en el orden correcto
meses_orden = ['enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 'diciembre']

# Creo un DataFrame con los meses en el orden correcto
df_mes_orden = pd.DataFrame(meses_orden, columns=['name_month'])

# Merge para asegurar que todos los meses estén presentes y con un conteo de 0 si no hay películas
df_resultado = df_mes_orden.merge(conteo_peliculas, on='name_month', how='left').fillna(0).astype({'total_movies': 'int'})

print("\nDataFrame Final con el Conteo de Películas por Mes:")
print(df_resultado)
```

```python
# Verifico si la suma de la columna peliculas totales coincide con la cantidad de filas (que coincide con cantidad de películas) y sabiendo que no hay valores nulos.
print(df_resultado['total_movies'].sum())
```

```python
# Noto que coincide con el total de filas.
```

```python
df_resultado
```

```python
# Almaceno el Dataframe en un archivo con formato Parquet y es el que utilizaremos para la primer consulta de la API

df_resultado.to_parquet('Peliculas por mes', engine='pyarrow')
```

```python
# Crearé un Dataframe que sirva para la segunda consulta de la API
```

```python
# Transformo la columna de fecha de estreno en formato datetime y luego obtengo el día de la semana en inglés.
dataset_movies['release_date'] = pd.to_datetime(dataset_movies['release_date'])
dataset_movies['release_day'] = dataset_movies['release_date'].dt.day_name()

# Diccionario para traducir días de la semana al español
traduccion_dias = {
    'Monday': 'lunes',
    'Tuesday': 'martes',
    'Wednesday': 'miercoles',
    'Thursday': 'jueves',
    'Friday': 'viernes',
    'Saturday': 'sabado',
    'Sunday': 'domingo'
}

# Traduzco el día de la semana al español
dataset_movies['release_day'] = dataset_movies['release_day'].map(traduccion_dias)

# Creo el nuevo DataFrame con las columnas deseadas
df_dia = dataset_movies[['title', 'release_day']]

print("\nNuevo DataFrame con Títulos y Días de la Semana en Español:")
print(df_dia)
```

```python
# Contaré el número de películas en cada día de la semana
conteo = df_dia.groupby('release_day').size().reset_index(name='total_peliculas')


# Lista de días de la semana en el orden correcto
dias_semana_orden = ['lunes', 'martes', 'miercoles', 'jueves', 'viernes', 'sabado', 'domingo']

#Creo un DataFrame con los días de la semana en el orden correcto
df_dias_orden = pd.DataFrame(dias_semana_orden, columns=['release_day'])

# Merge para asegurar que todos los días de la semana estén presentes y con un conteo de 0 si no hay películas
df_dia = df_dias_orden.merge(conteo, on='release_day', how='left').fillna(0).astype({'total_peliculas': 'int'})

print("\nDataFrame Final con el Conteo de Películas por Día de la Semana:")
print(df_dia)
```

```python
df_dia
```

```python
# Verifico que sea correcto el paso anterior. La suma del total de peliculas coincide con la cantidad de filas, es decir, con la cantidad total de peliculas.
df_dia['total_peliculas'].sum()
```

```python
# Almaceno el Dataframe en un archivo con formato Parquet y es el que utilizaremos para la segunda consulta de la API

df_dia.to_parquet('Peliculas por día', engine='pyarrow')
```
